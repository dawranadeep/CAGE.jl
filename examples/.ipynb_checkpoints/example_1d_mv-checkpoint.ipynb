#       Test the code of MKLE 

### Domain: [0,1]. Covariance kernel: Exponential with lengthscale 0.2

```julia
x = range(0, 1, 1000) |> collect;
C11 = exp.( -pairwise(Euclidean(), x, x)/5);
C22 = exp.( -pairwise(Euclidean(), x, x)/5);
C12 = 0.0 * C11;
```

### Use a Fourier basis for each individual process

```julia
B1 = generate_basis_1d(x, "Fourier"; num_args = 10);
B2 = generate_basis_1d(x, "Fourier"; num_args = 10);

# Orthonormalize the basis (Although originally orthonormal form)
Q1 = orthonormalize_basis_1d(x, B1);
Q2 = orthonormalize_basis_1d(x, B2);
```

### Compute the MKLE

(psi12, lam12, psi11, lam11, psi22, lam22) = compute_known_MKLE_1d(x, C11, C22, C12, Q1, Q2);

### Check they are eigenfunctions

```julia
tmp = [psi12[1] psi12[2]];
W2 = zeros(size(tmp,2), size(tmp,2));
for i in 1:size(tmp, 2)
    for j in 1:size(tmp, 2)
        W2[i,j] = trapz(x, tmp[:,i] .* tmp[:, j] );
        W2[j,i] = W2[i,j];
    end
end

round.(W2, digits=4)
```

### Plot individual process eigenfunctions

```julia
p1 = plot(x, tmp[:,1], legend = false);
for i in 2:Int(size(tmp,2)/2)
    p1 = plot!(x, tmp[:,i], legend = false);
end

p2 = plot(x, tmp[:, 1+ Int(size(tmp,2)/2)], legend = false);
for i in (2+ Int(size(tmp,2)/2)):size(tmp,2)
    p2 = plot!(x, tmp[:,i], legend = false);
end

plot(p1, p2, layout = (1, 2), legend = false)
```




### Run it in loop with increasing covariances



```julia
psi12_1s = zeros(size(psi12[1],1), size(psi12[1],2), 10);
psi12_2s = zeros(size(psi12[2],1), size(psi12[2],2), 10);

lam12s = zeros(size(lam12,1), 10);
for i in 1:10
    print(i, "\n");
    (tmp1, tmp2, ~, ~, ~, ~) = compute_known_MKLE_1d(x, C11, C22, (i - 1)/10 * C11, Q1, Q2);
    psi12_1s[:, :, i] = tmp1[1];
    psi12_2s[:, :, i] = tmp1[2];
    lam12s[:, i] = tmp2;
end
```

### Plot the squared eigenfunctions for each of the processes (square to remove confusion of sign)

```julia
pps = [];
for i in 1:10
    pp = plot(x, psi12_1s[:,  (end-5), i].^2);
    pp = plot!(x, psi12_2s[:, (end-5), i].^2);
    pp = title!(join(["Corr=", (i-1)/10]));
    push!(pps, pp);
end

plot(pps[1], pps[2], pps[3], pps[4], pps[5], pps[6], layout = (3, 2), legend = false)
```

### Compare the eigenvalues for each of the processes

```julia
pp = plot(lam12s[(end - 5):end, 1]);
for i in 2:10
    pp = plot!(lam12s[(end-5):end, i]);
end
pp
```
